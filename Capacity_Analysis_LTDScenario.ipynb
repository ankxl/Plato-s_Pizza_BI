```python
# Import necessary library
import numpy as np
import pandas as pd
from datetime import datetime, date, time, timedelta
import math
```

```python
# Read the dataset in to files
orders = pd.read_csv('C:\\Users\\ankit\\OneDrive\\Desktop\\Maven Pizza\\data\\orders.csv',parse_dates=[['date', 'time']])
order_details = pd.read_csv('C:\\Users\\ankit\\OneDrive\\Desktop\\Maven Pizza\\data\\order_details.csv')
pizzas = pd.read_csv('C:\\Users\\ankit\\OneDrive\\Desktop\\Maven Pizza\\data\\pizzas.csv')
pizza_types = pd.read_csv('C:\\Users\\ankit\\OneDrive\\Desktop\\Maven Pizza\\data\\pizza_types.csv',encoding= 'unicode_escape')
```

```python
# Merge the datasets Pizza and Pizza Types
pz_df = pizza_types.merge(pizzas,how='inner')
pz_df.info()
```

```python
# Merge the dataset Orders and Order Details
order_df = orders.merge(order_details,how='inner')
order_df.info()
```

```python
# Merge all the datasets in to mario_df
mario_df = order_df.merge(pz_df,how='inner')
mario_df.info()
```

```python
# Create a column for servings: How many people are sharing a Pizza
## Assumption:
## 1. Small Pizza is for single person consumption
## 2. Medium Pizza is shared between 2 people
## 3. Large Pizza is shared between 4 people
## 4. Xtra Large Pizza is shared between 6 people
## 5. Xtra Xtra Large Pizza is shared between 8 people
mario_df['serving'] = 1
mario_df.loc[mario_df['size']=='S'].head()
```

```python
mario_df['serving'] = np.where(mario_df['size'] == 'M',2,mario_df.serving)
mario_df.loc[mario_df['size']=='M'].head()
```

```python
mario_df['serving'] = np.where(mario_df['size'] == 'L',4,mario_df.serving)
mario_df.loc[mario_df['size']=='L'].head()
```

```python
mario_df['serving'] = np.where(mario_df['size'] == 'XL',6,mario_df.serving)
mario_df.loc[mario_df['size']=='XL'].head()
```

```python
mario_df['serving'] = np.where(mario_df['size'] == 'XXL',8,mario_df.serving)
mario_df.loc[mario_df['size']=='XXL'].head()
```

```python
# Calculate how many people were served in each order

mario_df['people_served'] = mario_df['quantity'] * mario_df['serving']
mario_df.head()
```

```python
# Calculate tables required for each order

mario_df['tables'] = mario_df['people_served'].apply(lambda x: math.ceil(x/4))
mario_df.head()
```

```python
# Calculate the revenue for each line item

mario_df['revenue']= mario_df.price * mario_df.quantity
mario_df.head()
```

```python
# Create a column for time when order is served and completed (40 mins after placing the order: 20 mins to serve the order
# and 20 mins to eat the ordered items. This is assuming seating was not an issue.)
mario_df['outer_end'] = mario_df['date_time'].apply(lambda x: datetime(x.year,x.month,x.day,
                              x.hour,x.minute,x.second) + timedelta(minutes=40))
```

```python
# Creating a series with all order date time slots 
order_datetime = mario_df['date_time']
order_datetime.sort_values()
order_datetime.shape
```

```python
# Creating a series with all order completed date time slots
served_datetime = mario_df['outer_end']
served_datetime.sort_values()
served_datetime.shape

```

```python
# Merge the series of order and order completed date time slots for capacity analysis
all_datetime = pd.concat([order_datetime,served_datetime],axis=0,ignore_index=True,sort=True)
all_datetime.sort_values(axis=0,ascending=True,inplace=True,ignore_index=True)
all_datetime.shape
```

```python
# Series to DataFrame
all_datetime = pd.DataFrame(all_datetime,columns=['date_time'])
all_datetime['date'] = pd.to_datetime(all_datetime['date_time']).dt.date
all_datetime.drop_duplicates(inplace=True,ignore_index=True)
# all_datetime
```

```python
# Dataset to store the final capacity snapshot
capacity_date = pd.DataFrame(pd.date_range(start='01/01/2015',end='31/12/2015'),columns=['Date'])
capacity_date['Overflow_tables']=0
capacity_date['Overflow_seats']=0
capacity_date.Date=pd.to_datetime(capacity_date.Date).dt.date
capacity_date.tail()
```

```python
# Capacity Analysis
# - For each date where an order was placed identify all the unique order or order completed timeslots
# - For each order or order completed timeslot on that date identify the number of tables and seats required to serve 
# the order or tables/ seats released when order is completed.
# - If the tables occupied is greater than 15 or is the seats occupied is greater than 60 flag capacity exceeded for
# tables/seats by 1
# - At the end of the inner loop i.e. when all order or order completed timeslots are traversed then append the value of
# tables/ seat capacity violation to the final dataframe

for i in capacity_date.Date:
    df = all_datetime.loc[all_datetime['date']==i]
    tables_occupied = 0
    seats_occupied = 0
    overflow_tables = 0
    seats_occupied = 0
    for index, row in df.iterrows():
        
        temp_df = mario_df.loc[mario_df['date_time']==row['date_time']]
        if not temp_df.empty:
            tables_occupied = tables_occupied + temp_df.tables.sum()
            seats_occupied = seats_occupied + temp_df.people_served.sum()
        
        temp_df = mario_df.loc[mario_df['outer_end']==row['date_time']]
        if not temp_df.empty:
            tables_occupied = tables_occupied - temp_df.tables.sum()
            seats_occupied = seats_occupied - temp_df.people_served.sum()
        
        if tables_occupied > 15:
            overflow_tables = overflow_tables + 1
        
        if seats_occupied > 60:
            seats_occupied = seats_occupied + 1
        
    capacity_date.loc[capacity_date['Date'] == i,'Overflow_tables'] = overflow_tables
    capacity_date.loc[capacity_date.Date == i,'Overflow_seats'] = seats_occupied
```
