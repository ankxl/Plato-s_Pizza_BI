```python
# Import necessary library
import numpy as np
import pandas as pd
from datetime import datetime, date, time, timedelta
import math
```

```python
# Read the dataset in to files
orders = pd.read_csv('C:\\Users\\ankit\\OneDrive\\Desktop\\Maven Pizza\\data\\orders.csv',parse_dates=[['date', 'time']])
order_details = pd.read_csv('C:\\Users\\ankit\\OneDrive\\Desktop\\Maven Pizza\\data\\order_details.csv')
pizzas = pd.read_csv('C:\\Users\\ankit\\OneDrive\\Desktop\\Maven Pizza\\data\\pizzas.csv')
pizza_types = pd.read_csv('C:\\Users\\ankit\\OneDrive\\Desktop\\Maven Pizza\\data\\pizza_types.csv',encoding= 'unicode_escape')
```

```python
# Merge the datasets Pizza and Pizza Types
pz_df = pizza_types.merge(pizzas,how='inner')
pz_df.info()
```

```python
# Merge the dataset Orders and Order Details
order_df = orders.merge(order_details,how='inner')
order_df.info()
```

```python
# Merge all the datasets in to mario_df
mario_df = order_df.merge(pz_df,how='inner')
mario_df.info()
```

```python
# Create a column for servings: How many people are sharing a Pizza
## Assumption:
## 1. Small Pizza is for single person consumption
## 2. Medium Pizza is shared between 2 people
## 3. Large Pizza is shared between 4 people
## 4. Xtra Large Pizza is shared between 6 people
## 5. Xtra Xtra Large Pizza is shared between 8 people
mario_df['serving'] = 1
mario_df.loc[mario_df['size']=='S'].head()
```

```python
mario_df['serving'] = np.where(mario_df['size'] == 'M',2,mario_df.serving)
mario_df.loc[mario_df['size']=='M'].head()
```

```python
mario_df['serving'] = np.where(mario_df['size'] == 'L',4,mario_df.serving)
mario_df.loc[mario_df['size']=='L'].head()
```

```python
mario_df['serving'] = np.where(mario_df['size'] == 'XL',6,mario_df.serving)
mario_df.loc[mario_df['size']=='XL'].head()
```

```python
mario_df['serving'] = np.where(mario_df['size'] == 'XXL',8,mario_df.serving)
mario_df.loc[mario_df['size']=='XXL'].head()
```

```python
# Calculate how many people were served in each order

mario_df['people_served'] = mario_df['quantity'] * mario_df['serving']
mario_df.head()
```

```python
# Calculate tables required for each order

mario_df['tables'] = mario_df['people_served'].apply(lambda x: math.ceil(x/4))
mario_df.head()
```

```python
# Calculate the revenue for each line item

mario_df['revenue']= mario_df.price * mario_df.quantity
mario_df.head()
```

```python
mario_df.groupby('order_id').revenue.sum()
```

```python
# Create a column for time when order is served and completed (40 mins after placing the order: 20 mins to serve the order
# and 20 mins to eat the ordered items. This is assuming seating was not an issue.)
mario_df['outer_end'] = mario_df['date_time'].apply(lambda x: datetime(x.year,x.month,x.day,
                              x.hour,x.minute,x.second) + timedelta(minutes=40))
```

```python
dates = pd.to_datetime(mario_df['date_time']).dt.date.unique()
dates.shape
```

```python
dates = pd.DataFrame(dates,columns=['Date'])
dates.head()
dates.sort_values(by='Date',axis=0,inplace=True,ignore_index=True,ascending=True)
dates['Overflow_tables']=0
dates['Overflow_seats']=0
dates.head()
```

```python
for i in dates.Date:
    print(i)
    hour = 0
    minute = 0
    second = 0
    overflow_tables = 0
    overflow_seat = 0
    tables_occupied = 0
    seats_occupied = 0
    for hour in range(0,24):
        for minute in range(0,60):
            for second in range(0,60):
                flag = 0
                time_ = time(hour,minute,second)
                dt = datetime.combine(i,time_)
                
                df = mario_df.loc[mario_df['date_time'] == dt]
                if not df.empty:
                    flag = 1
                    tables_occupied = tables_occupied + df.tables.sum()
                    seats_occupied = seats_occupied + df.people_served.sum()
                
                df = mario_df.loc[mario_df['outer_end'] == dt]
                if not df.empty:
                    tables_occupied = tables_occupied - df.tables.sum()
                    seats_occupied = seats_occupied - df.people_served.sum()
                
                if tables_occupied > 15 and flag != 0:
                    overflow_tables = overflow_tables + 1
                
                if seats_occupied > 60 and flag != 0:
                    overflow_seat = overflow_seat + 1
    dates.loc[dates['Date'] == i,'Overflow_tables'] = overflow_tables
    dates.loc[dates.Date == i,'Overflow_seats'] = overflow_seat
```
